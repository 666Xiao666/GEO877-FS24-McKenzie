I calculate the haversine distances [m] of the bicycle trips, based on their starting and ending coordinates. Base of code from course GEO877 - Spatial Analysis, FS2024

```python
import pandas as pd
import os, sys
import numpy as np
import math

#reads in csv sources as df
path = "CoGo_Bikerental_colorado_US/cleaned_data/"
filelist = os.listdir(path)
print(filelist)

for f in filelist:
    source = path + f
    df = pd.read_csv(source)

    #conversion of lat and lng values into radians
    lat1 = np.radians(df["start_lat"])
    lat2 = np.radians(df["end_lat"])
    lng1 = np.radians(df["start_lng"])
    lng2 = np.radians(df["end_lng"])

    #calculation of haversine distances and conversion into meters
    dlat = lat1 - lat2
    dlng = lng1 - lng2
    dista = np.sin(dlat/2)**2 + np.cos(lng1) * np.cos(lng2) * np.sin(dlng/2)**2
    distc = 2 * np.arcsin(np.sqrt(dista))  
    dist = round(6367000 * distc,2)

    #append haversine distances to datasets
    df["haversine_distance [m]"] = dist


    #for checking on output
    display(df.head())

    #storing cleaned data
    output = "CoGo_Bikerental_colorado_US/cleaned_data/"
    storage = output + f 
    df.to_csv(storage)
```
